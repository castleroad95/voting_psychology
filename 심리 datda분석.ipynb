## Data 분석하기

```python
import pandas as pd

train_file = pd.read_csv('C:/Users/castl/Desktop/2020_job_academy/project/voting_psychology/input/train.csv')
train_file.head()
```

```python
# columns info 분석

train_file.info()
```

```python
# 투표한 사람중 성별, 인종, 종교별 알아보기

train_file.groupby(['gender','voted'])['voted'].count()
train_file.groupby(['race','voted'])['voted'].count()
train_file.groupby(['religion','voted'])['voted'].count()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='gender',y='voted',data=train_file)
plt.show()

sns.barplot(x='race',y='voted',data=train_file)
plt.xticks(rotation=45)
plt.show()

sns.barplot(x='religion',y='voted',data=train_file)
plt.xticks(rotation=90)
plt.show()
```
